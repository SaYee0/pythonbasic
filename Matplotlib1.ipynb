```python
import numpy as np
x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
plt.title("x축과 y축의 tick label 설정")
plt.plot(x,c)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1],["Low","Zero","High"])
plt.show()
```

```python
x1 = np.linspace(0.0,5.0)
x2 = np.linspace(0.0,2.0)
y1 = np.cos(2*np.pi*x1) * np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

ax1 = plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2,1,2)
plt.plot(x2,y2,'r.-')
plt.xlabel('time(s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic

a = titanic.alive.value_counts()['no']
b = titanic.alive.value_counts()['yes']
num = [a,b]
alive = ["사망자","생존자"]
colors = ["red","green"]
plt.title("타이타닉 생존 여부")
plt.xlabel("인원 수")
plt.barh(alive,num, color=colors)
```

```python
labels = ['월세','전세','자가','기타']
sizes = [15,30,45,10]
colors = ['yellowgreen','gold','lightskyblue','lightcoral']
explode = [0,0,0.5,0]
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
       autopct='%1.1f%%', shadow=True, startangle=50)
plt.axis('equal')
plt.show()
```

```python
import pandas as pd

bins = [0, 20, 30, 50, 70, 100]
label = ['미성년자','청년','장년','중년','노년']


age = pd.cut(titanic['age'].dropna(), bins, labels=label)
print(age)
a=age.value_counts()['미성년자']
b=age.value_counts()['청년']
c=age.value_counts()['장년']
d=age.value_counts()['중년']
e=age.value_counts()['노년']

nums = [a,b,c,d,e]
labels = ["미성년자", "청년", "장년", "중년", "노년"]
explode = (0.1, 0, 0, 0, 0)
colors = ["yellowgreen", "gold", "lightskyblue", "lightcoral", "orange"]

plt.pie(nums, labels=labels, explode=explode, colors=colors,
       autopct='%1.1f%%', shadow=True, startangle = 120)
```
