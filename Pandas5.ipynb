```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
```

```python
titanic
```

```python
titanic.dtypes
```

```python
def change_boolean(x):
    if x["alive"] == "yes":
        return True
    else:
        return False
    
titanic["alive"] = titanic.apply(change_boolean, axis =1)
```

```python
titanic["alive"] = titanic.apply([lambda x: True if x["alive"]=="yes" else ])
```
