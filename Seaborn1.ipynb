```python
import matplotlib as mpl
# Add every font at the specified location
font_dir = ['./fonts']
for font in mpl.font_manager.findSystemFonts(fontpaths=font_dir):
    print(mpl.font_manager.FontProperties(fname=font).get_name())
    mpl.font_manager.fontManager.addfont(font)
```

```python
import pandas as pd
import numpy as np
pd.pivot_table(titanic, "survived", index="class", columns="who")
```

```python
pd.pivot_table(titanic, "survived", index="class", columns="deck", aggfunc=np.mean)
```

```python
pd.pivot_table(titanic, "survived", index="class", columns="deck", aggfunc=np.mean, fill_value=0)
```

```python
a = pd.pivot_table(titanic, "survived", index="class", columns="who")
sns.heatmap(a, cmap='crest')
```

```python
sns.heatmap(a, cmap='mako', annot=True, fmt='.2f', linewidth=.5)
```
